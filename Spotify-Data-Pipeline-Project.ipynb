### Extracting data from spotify using sp_object & passing play_list_id to 'playlist_tracks' function

```python
data = sp_object.playlist_tracks(play_link_id)
```

```python
data
```

### Extracting required information about songs, artist & albums.

```python
# 1st album

#Use json formator to view data 

album = []

for row in data['items']:
    album_id = row['track']['album']["id"]
    album_name = row['track']['album']["name"]
    album_release_data = row['track']['album']["release_date"]
    album_total_track = row['track']['album']["total_tracks"] 
    
    album_elements = {'album_id':album_id,
                     'album_name':album_name,
                     'album_release_data':album_release_data,
                     'album_total_track':album_total_track}
    album.append(album_elements)
    print(album)
    
    

```

```python
# convering list into DataFram

import pandas as pd
album_df = pd.DataFrame(album)
album_df.head()
```

```python
# 2nd song

song = []
for row in data['items']:
    song_id = row['track']['id']
    song_name = row['track']['name']
    song_duration = row['track']['duration_ms']
    song_url = row['track']['external_urls']['spotify']
    song_popularity = row['track']['popularity']
    song_added = row['added_at']
    album_id = row['track']['album']['id']
    artist_id = row['track']['album']['artists'][0]['id']
    song_element = {'song_id':song_id,'song_name':song_name,'duration_ms':song_duration,'url':song_url,
                    'popularity':song_popularity,'song_added':song_added,'album_id':album_id,
                    'artist_id':artist_id
                   }
    song.append(song_element)
    
```

```python
song_df = pd.DataFrame(song) 
song_df.head()
```

```python
#3rd artist

artist_list = []
for row in data['items']:
    for key, value in row.items():
        if key == "track":
            for artist in value['artists']:
                artist_dict = {'artist_id':artist['id'], 'artist_name':artist['name'], 'external_url': artist['href']}
                artist_list.append(artist_dict)
```

```python
artist_df = pd.DataFrame(artist_list)
artist_df.head()
```
